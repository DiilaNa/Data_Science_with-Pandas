```python
import pandas as pd
from unicodedata import numeric
```

```python
df_car_sale = pd.read_csv("../Assets/Car_sales.csv",engine="python")
```

```python
df_car_sale
```

```python
df_car_sale = df_car_sale[['Manufacturer','Sales_in_thousands','Vehicle_type','Price_in_thousands','Engine_size','Horsepower','Fuel_capacity']]
```

```python
df_car_sale
```

```python
# calculate the number of unique values
df_car_sale.nunique()
```

```python
# agg()
# calculate the sum of columns
df_car_sale.agg('sum',numeric_only=True)
```

```python
# calculate the mean of numeric columns
df_car_sale.agg('mean',numeric_only=True)
```

```python
# calculate the mean and count values
numeric_cols = df_car_sale.select_dtypes(include='number')
numeric_cols.agg(['mean','count'])
```

```python
df_car_sale
```

```python
# different aggregations per columns , sum and mean of 'sales_in_thousands' and sum and max of 'Price_in_thousands'
df_car_sale.agg({'Sales_in_thousands':['sum','mean'],
                 'Price_in_thousands':['sum','max']})
```

```python
df_car_sale
```

```python
# aggregate over the columns (sum both columns row vise)
df_car_sale[['Sales_in_thousands','Price_in_thousands']].agg('sum',axis=1)
```

```python
# aggregate different functions and rename the index of resulting data
df_car_sale.agg( x = ('Sales_in_thousands','sum'),
                 y = ('Price_in_thousands','sum'))
```

```python
#-----------------The Split-Apply-Combine Strategy-------------------------
# split data into separate groups
passenger_filter = df_car_sale['Vehicle_type'] == 'Passenger'
car_filter = df_car_sale['Vehicle_type'] == 'Car'

#df_car_sale['Vehicle_type'].value_counts()
```

```python
# apply an operation
passenger_average = df_car_sale[passenger_filter]['Sales_in_thousands'].mean()
car_average = df_car_sale[car_filter]['Sales_in_thousands'].mean()
```

```python
pd.DataFrame({'Vehicle_type':['Car','Passenger'],
              'Sales_in_thousands':[car_average,passenger_average]})
```

```python
#--------------------------GroupBy()-------------------------
# group by vehicle_type and calculate the mean add as_index=False to remove vehicle type as index

df_car_sale.groupby('Vehicle_type').mean(numeric_only=True)
```

```python
# Group by manufacturer - get a group by object when use group by
groupBy_obj = df_car_sale.groupby('Manufacturer')
```

```python
groupBy_obj
```

```python
# groups attributes and keys - index numbers are same
groupBy_obj.groups
```

```python
groupBy_obj.groups.keys()
```

```python
# group by manufacturer and calculate the mean
df_car_sale.groupby('Manufacturer').mean(numeric_only=True)
```

```python
# group by manufacturer and calculate the sum
df_car_sale.groupby('Manufacturer').sum(numeric_only=True)
```

```python
# group by Vehicle_type and count values (Nan values are not counted with count())
df_car_sale.groupby('Vehicle_type').count()
```

```python
# find null values in columns
df_car_sale.isnull().sum()
```

```python
# group by engine size and count values without dropping NaN Values
df_car_sale.groupby('Engine_size',dropna=False).count()
```

```python
#----------------GroupBy() and agg() -------------------------

# find the minimum and maximum values on each column
df_car_sale.groupby('Vehicle_type').agg(['min','max'])
```

```python
# set a name for aggregated column and
# find the min Engine_Size and max Horsepower
df_car_sale.groupby('Vehicle_type').agg(min_engine_size=('Engine_size','min'),max_horsepower=('Horsepower','max'))
```

```python
# calculate the sum of Sales_in_thousands and mena 0f Price_in_thousands
df_car_sale.groupby('Manufacturer').agg(sum_sales=('Sales_in_thousands','sum'),mean_price=('Price_in_thousands','mean'))
```

```python
# same above cell activity but without agg()
#df_car_sale.groupby('Manufacturer').sum()['Sales_in_thousands']
df_car_sale.groupby('Manufacturer').mean(numeric_only=True)['Price_in_thousands']
```

```python
# -- Lambda Function --
df_car_sale.groupby('Manufacturer').sum(numeric_only=True)
```

```python
# get actual sales and price by values
df_car_sale.groupby('Manufacturer').sum().apply(lambda x:x*1000)[['Price_in_thousands','Sales_in_thousands']]
```

```python
#--------------Filter () ----------------------
df_car_sale.mean(numeric_only=True)
```

```python
# get sum of the sales are more than 52
def filter_func(x):
    return x['Sales_in_thousands'].sum()> 52

df_filter = df_car_sale.groupby(['Manufacturer']).filter(filter_func)
df_filter
```
