```python
#------------------------Page Five----------------------------
import pandas as pd
import numpy as np
```

```python
#-----------------------Joins----------------------------#
#-------Inner Join------
# Merge
df1 = pd.DataFrame({'id':['A','B','C','D'],
                    'age':[21,22,23,24]})
df2 = pd.DataFrame({'id':['C','D','E','F'],
                    'job':['Doctor','Accountant','Developer','Teacher']})
```

```python
print(df1)
print(df2)
```

```python
# Inner Join - Example one
df_inner_join = df1.merge(df2, on='id',how='inner')

print(df_inner_join)
```

```python
# shape
print(df1.shape)
print(df2.shape)
print(df_inner_join.shape)
```

```python
df_movies = pd.read_csv('../Assets/IMDb movies.csv')
df_ratings = pd.read_csv('../Assets/IMDb ratings.csv')
```

```python
df_movies = df_movies[['imdb_title_id', 'title', 'year', 'genre', 'country']]
df_ratings = df_ratings[['imdb_title_id', 'total_votes', 'mean_vote']]
```

```python
# Example Two
df_movies_inner = df_movies.merge(df_ratings, on='imdb_title_id', how='inner')
```

```python
print(df_movies_inner.shape)
print(df_ratings.shape)
print(df_movies.shape)
```

```python
df_movies_inner
```

```python
# Inner join with pd.merge
pd.merge(df_movies_inner, df_movies, on='imdb_title_id', how='inner')
```

```python
#----------------------Full Join-----------------------------------
df1.merge(df2,on='id',how='outer')
```

```python
df_movies_outer = df_movies.merge(df_ratings,on='imdb_title_id',how='outer')
```

```python
print(df_movies.shape)
print(df_ratings.shape)
print(df_movies_outer.shape)
```

```python
pd.merge(df_movies,df_ratings,on='imdb_title_id',how='outer')
```

```python
#-------------Exclusive Full join(show result opposite to inner join)-------------------
df1.merge(df2,on='id',how='outer',indicator=True)
```

```python
# query - select left only and right only values
df1.merge(df2,on='id',how='outer',indicator=True).query("_merge=='left_only' or _merge=='right_only'")
```

```python
df_movies_exclusive = df_movies.merge(df_ratings,on='imdb_title_id',how='outer',indicator=True)
```

```python
df_movies_exclusive
```

```python
pd.merge(df_movies,df_ratings,on='imdb_title_id',how='outer',indicator=True).query("_merge=='left_only' or _merge=='right_only'")
```

```python
#-------------------------Left Join--------------------------
df1.merge(df2,on='id',how='left')
```

```python
# extract 50% from df movies
df_movies_sample = df_movies.sample(frac=0.5)
```

```python
df_left = df_movies_sample.merge(df_ratings,on='imdb_title_id',how='left')
```

```python
df_left
```

```python
print(df_movies_sample.shape)
print(df_movies.shape)
print(df_ratings.shape)
print(df_left.shape)
```

```python
pd.merge(df_movies_sample,df_ratings,on='imdb_title_id',how='left')
```

```python
#--------------Exclusive Left Join (result of the left join but without common values to both , belongs only to left)--------------------------

# when in an exclusive join always how parameter is outer
df1.merge(df2,on='id',how='outer',indicator=True).query("_merge=='left_only'")
```

```python
# make a copy of df_movies
df_movies_2 = df_movies.copy()
```

```python
# set the first 1000 values of 'imdb_title_id' column as 'tt1234567890'
for index in df_movies_2.index:
    if index<1000:
        df_movies_2.loc[index,'imdb_title_id'] = 'tt1234567890'
```

```python
# merge df_movies_2  and df_ratings (exclusive _left join)
df_exclusive_left = df_movies_2.merge(df_ratings,on='imdb_title_id',how='outer',indicator=True).query("_merge=='left_only'")
```

```python
# shape
print(df_exclusive_left.shape)
print(df_movies_2.shape)
print(df_ratings.shape)
```
