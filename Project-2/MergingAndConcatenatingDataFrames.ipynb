```python
import pandas as pd
```

```python
# -----------------------Concatenate - Concat()-----------------
# -----------concat vertically - (axis=0) default----------------
df1 = pd.DataFrame({'id':['A','B','C'],
                    'age':[30,50,60]})
df2 = pd.DataFrame({'id':['D','E','F'],
                    'age':[34,52,45]})
```

```python
pd.concat([df1,df2],axis=0,ignore_index=True)
```
