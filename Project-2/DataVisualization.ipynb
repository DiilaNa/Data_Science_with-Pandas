```python
import pandas as pd
import numpy as np
from matplotlib.pyplot import xlabel
```

```python
df_population_raw = pd.read_csv('../Assets/population.csv')
```

```python
df_population_raw
```

```python
# making Pivot table

# dropping null values
df_population_raw.dropna(inplace=True)
```

```python
# making a pivot table
df_pivot = df_population_raw.pivot(index='Year', columns='Country Name', values='Value')
```

```python
df_pivot
```

```python
# select some countries
df_pivot = df_pivot[['United States','India','China','Brazil','Indonesia']]
```

```python
# show pivot table
df_pivot
```

```python
#----------------Visualizations---------------------
#Lineplot - !pip install matplotlib
import matplotlib.pyplot as plt
df_pivot.plot(kind='line', xlabel='Year', ylabel='Population', title='Population by Year (1960-2025)',figsize=(8,4))
plt.show()
```

```python
#Barplot

#-single bar plot
# selecting only one year
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
```

```python
df_pivot_2020
```

```python
# transform df from row to column or vice versa
df_pivot_2020 = df_pivot_2020.T
```

```python
#making the barplot
df_pivot_2020.plot(kind='bar',color='orange',xlabel='Country Name',ylabel='Population',title='Population by Year (2020-2021)',figsize=(8,4))
plt.show()
```

```python
# barplot group by 'n' variables

# selecting few years
df_pivot_sample = df_pivot[df_pivot.index.isin([1980,1990,2000,2010,2020])]
```

```python
df_pivot_sample.plot(kind='bar',xlabel='Year',ylabel='Population',title='Population by Year (2020-2021)',figsize=(8,4))
```
