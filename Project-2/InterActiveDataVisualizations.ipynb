```python
import pandas as pd
import plotly.express as px
```

```python
df_population_raw = pd.read_csv("../Assets/population.csv")
```

```python
df_population_raw.dropna(inplace=True)
```

```python
df_pivot = df_population_raw.pivot(index='Year', columns='Country Name', values='Value')
```

```python
df_pivot = df_pivot[['United States','India','China','Indonesia','Brazil']]
```

```python
df_pivot
```

```python
df_pivot.iplot(kind = 'bar')
# works only in notebook as i am using pycharm i need to use above codes
```

```python
# interactive Line chart
fig = px.line(df_pivot, x=df_pivot.index, y=df_pivot.columns)
fig.show()
```

```python
# interactive barplot

df_pivot_2020=df_pivot[df_pivot.index.isin([2020])]
```

```python
df_pivot_2020 = df_pivot_2020.T
```

```python
df_pivot_2020
```

```python
fig = px.bar(df_pivot_2020, x=df_pivot_2020.index, y=df_pivot_2020.columns,title='Population by Country',color_discrete_sequence=px.colors.sequential.Darkmint)
fig.show()
```

```python
# Interactive barplot grouped by 'n' variables
df_pivot_sample = df_pivot[df_pivot.index.isin([1970,1980,1990,2000,2010,2020])]
```

```python
df_pivot_sample
```

```python
fig = px.bar(df_pivot_sample, x=df_pivot_sample.index, y=df_pivot_sample.columns,title='Population[1970-2020]')
fig.show()
```

```python
#-Interactive Pie chart

df_pivot_2020.rename(columns={2020:'2020'},inplace=True)
```

```python
df_pivot_2020 = df_pivot_2020.reset_index()
```

```python
fig = px.pie(df_pivot_2020, values='2020',names='Country Name',title='2020')
fig.update_traces(textinfo='percent', textposition='inside')
fig.show()
```

```python
# Boxplot
import plotly.express as px

fig = px.box(y=df_pivot['United States'], title='United States', labels={'y': 'Values', 'x': 'United States'})
fig.show()
```

```python
fig = px.box(df_pivot, title='Population by Country', labels={'value': 'Population', 'variable': 'Country'})
fig.show()
```

```python
# Interactive Histogram
fig = px.histogram(df_pivot['United States'], title='Population by Country', labels={'value': 'Population'}, nbins=3)
fig.show()
```

```python
fig = px.histogram(df_pivot[['United States','Indonesia']], title='Population by Country', labels={'value': 'Population'}, nbins=3)
fig.show()
```

```python
# Interactive ScatterPlot
fig = px.scatter(df_pivot)
fig.show()
```
