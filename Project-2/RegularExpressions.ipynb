```python
#-----------------------Page Six---------------
import re
```

```python
# Metacharacters

#1) \d          -   Digits(0-9)

#2) \D          -   No Digits(get all that are not digits)

#3) \w          -   Word Character (a-z,A-Z,_,0-9)

#4) \W          -   Not Word Characters(!,spaces,.,-)

#5) \s          -   Whitespace(space,tab-space)

#6) \S          -   No Whitespace

#7) .           -   All characters except new line

#8) \           -   Ignores any special characters except newlines

#8) ^           -   Beginning of a string

#9) $           -   End of a string
```

```python
#--------------Quantifiers----------------------------

# example -> Python!!!!!!!

# *     - Python!* -> will mak all exclamation marks in the word python(greedy)

# +     - Python!* -> capture one or more marks in the word,similar to *(greedy) get the longest string

# ?     - Python!? -> capture the shortest possible string (lazy)

# {3}   - Python!{2} -> capture 2 exclamation marks in the word (as the example)

# {n,}     -> capture n numbers of exclamation marks in the word (as the example)

# {3,4}    -> capture a range of exclamation marks in the word (as the example)

# ()    - capture specific group of characters

# []    - matches characters inside brackets

# [^]   - matches characters that are not inside brackets

# |     - Or , don't add or inside [] as it already splits, [89] means 8 or 9

# \b    - word boundary

# \B    - no word boundary

# \1    - Repeat what was matched in the first () group.Matches “hihi”.(hi)\1,
```

```python
#---------Search()--------------
text = """"Hello world!.
I love Python!!!!!
My first lucky number is 987-654-321
My second lucky number is 876-543-210
My third lucky number is 765-432-100
My fourth lucky number is 123-456-123-123-456
"""

re.search(r'\d',text)
```

```python
# flags
re.search(r'^hello',text,flags=re.IGNORECASE)
```

```python
#-------------------FindAll()------------
# exact numbers inside all the text variable
re.findall(r'\d',text)
```

```python
# Match only punctuation
re.findall(r'[^\w\s]',text)
```

```python
# Match right date format (MM-DD-YYYY)
text = ''''
13-04-2021
2021-13-04
2021-04-13
'''
re.findall(r'\d{2}-\d{2}-\d{4}',text)
```

```python
# Match right username format (4-14 characters and only letters or digit)
text = '''
username10
abc
10
'''
re.findall(r'[a-zA-Z0-9]{4,14}',text)
```
