```python
# Identify missing data
df_netflix.isnull().sum().sort_values(ascending=False)
```

```python
df_netflix.columns
```

```python
# % of missing in each column
for column in df_netflix.columns:
    percentage = df_netflix[column].isnull().mean()
    print(column +' :' + str(round(percentage*100,2)))
```

```python
# Dealing with Missing Data

# remove column or row with .drop m , .dropna or .isnull

# drop column
df_netflix.drop('director',axis=1)
```

```python
# drop row
no_director = df_netflix[df_netflix['director'].isnull()].index
```

```python
df_netflix.drop(no_director,axis=0)
```

```python
df_netflix.drop(no_director,axis=0).isnull().sum()
```

```python
# drop row with isnull method

# get not null values - incase drop nulls
df_netflix[~(df_netflix['director'].isnull())]
```

```python
df_netflix[~(df_netflix['director'].isnull())].isnull().sum()
```

```python
# drop rows with dropna()
df_netflix.dropna(subset=['director'])
```

```python
df_netflix.dropna(subset=['director']).isnull().sum()
```

```python
# Use fillna() to replace NAN by the mean, median(numeric data) or mode (categorical data)

mode =''.join(df_netflix['rating'].mode())
df_netflix['rating'].fillna(mode,inplace=True)
```

```python
df_netflix.isnull().mean()
```

```python
# Replace it by an arbitrary number with fillna()
df_netflix['duration'].fillna('0',inplace=True)

# as an error comes up when converting obj to int we have to add 0 as and string here,check down astype method cell
```

```python
# replace the null value with backward -data in  1 column to second & forward -data in third column to second
df_netflix['director'].bfill()
```

```python
df_netflix['director'].ffill()
```

```python
# extracting data from "duration" column with split() and extract()

df_movie = df_netflix[df_netflix['type']=='Movie']

df_movie['duration'].str.split(expand=True)[0]
```

```python
# turn object to int using astype
df_movie['minute'] = df_movie['duration'].str.split(expand=True)[0].astype(int)

# warning comes as i rename the column name
```

----Identifying Outliers------(Not Common elements in a dataframe)

Ex: One student got 5 marks when everyone else got between 60–90

```python
# Using histograms to identify outliers within numeric data

# make histogram with pandas

df_movie['minute'].plot(kind='hist',bins=10)
```

```python
# get bar ranges with valu_counts method and numbr of bins
df_movie['minute'].value_counts(bins=10).sort_index()
```

```python
# Dealing with outliers
df_movie[(df_movie['minute']>31) &  (df_movie['minute']<218)]
```

```python
# Using boxplot to identify outliers within numeric data

# make a boxplot with pandas
df_movie['minute'].plot(kind='box',vert=False,color='blue',figsize=(10,5))
```

## IQR = Q3-Q1
## Q1 - 1.5*IQR
## Q2 + 1.5*IQR
