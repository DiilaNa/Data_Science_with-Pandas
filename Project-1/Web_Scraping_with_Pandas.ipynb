```python
# Read a .csv from multiple Urls - Multiple leagues

# root variable
root = "https://www.football-data.co.uk/mmz4281/"
```

```python
# creating a list of leagues
leagues = ['E0','E1','E2','E3','EC']
frames = []

# looping through leagues , read multiple csv and append it into a list..
for league in leagues:
   df = pd.read_csv(root + '2526' + '/' + league + '.csv')
   frames.append(df)
```

```python
#__________________Multiple Seasons____________________
# creating a list of leagues
leagues = ['E0','E2','E3']
frames = []

# looping through range of seasons and insert new column
for league in leagues:
    for season in range(15, 21):
        df = pd.read_csv(root + str(season) + str(season+1) + '/' + league + '.csv')
        df.insert(1, 'season', season)
        frames.append(df)
```

```python
# total frames (6 seasons * 3 leagues)
len(frames)
```
