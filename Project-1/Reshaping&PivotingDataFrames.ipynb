```python
#------------------------Page Four----------------------------
import pandas as pd
import numpy as np
df_gdp = pd.read_csv("../Assets/gdp.csv" ,engine="python")
df_supermarket = pd.read_excel("../Assets/SupermaketSales.xlsx")
```

```python
#--------------------Pivot Tables------------------------
# 1) pivot() - Reshape data based on column values (does not support data aggregation)

# 2) pivot_tables() - create a spreadsheet-style pivot table (support data aggregation)
```

```python
#-------------pivot()--------------
df_gdp.pivot(index="Year", columns="Country Name", values="Value")
```

```python
# ---------------------Pivot Table()----------------
df_supermarket['Date'] = df_supermarket['Date'].astype(str)
df_supermarket['Time'] = df_supermarket['Time'].astype(str)
```

```python
df_supermarket.pivot_table(index="Gender",
                        values=["Quantity","Total"],  # your numeric columns here
                       aggfunc="sum")
```

```python
# create a pivot table that shows how much male and female spent in each category
df_supermarket.pivot_table(index="Gender",
                           columns="Product line",
                       values="Total",  # your numeric columns here
                       aggfunc="sum")
```
