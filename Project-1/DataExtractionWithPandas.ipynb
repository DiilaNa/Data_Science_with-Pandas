```python
#------------------------Page Three----------------------------
import pandas as pd
import numpy as np
df_players = pd.read_csv("../Assets/fifa_players.csv", engine='python')
```

```python
# selecting data with loc() and iloc()

# loc() is label-based so we have to specify rows and columns based in their index/row and column labels

# Value---------df.loc['zero']
# List----------df.loc[['zero','two']]
# Slicing-------df.loc['zero':'two']

#-------------------------------------------------------------------------
# iloc() is integer position-based so we have to specify rows and columns by their integer position values (0-based integer position)

# Value-------- df.iloc[0]
# List----------df.iloc[[0,2]]
# Slicing-------df.iloc[0:2]
```
