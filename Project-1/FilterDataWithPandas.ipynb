```python
#------------------------Page Two----------------------------
import pandas as pd
import numpy as np
df_laptops = pd.read_csv("../Assets/laptop_data.csv", engine='python')
```

```python
df_laptops.head(3)
```

```python
#----------------------Filter dataframe based on a condition--------------------
# find which rows have "Apple" in the Company Column

df_laptops['Company'] == "Apple"

# this return values as boolean
```

```python
# filter dataframe based on a condition
df_laptops[df_laptops['Company'] == 'Apple']
```

```python
# get a count of apple companies
df_laptops[df_laptops['Company'] == 'Apple'].value_counts('Company')
```

```python
# which rows dont have "HP"
df_laptops[df_laptops['Company'] != 'HP'].value_counts('Company')
```

```python
# find laptops over 120 000
df_laptops[df_laptops['Price']>120000].value_counts('Company')
```

```python
#------------------------Create a column based on one condition (Where)-----------------------
# create an array based on price tiers, like if condition here
np.where(df_laptops['Price']>120000,'Expensive','Cheap')
```

```python
# set a new column
df_laptops['Price_Tier'] = np.where(df_laptops['Price']>120000,'Expensive','Cheap')
```

```python
# show first five rows
df_laptops.head(5)
```

```python
# get a count in price tier column
df_laptops.value_counts('Price_Tier')
```

```python
#----------------Filter dataframe based on two conditions (And and Or Operators)------------------
# condition
(df_laptops['Company'] == 'Apple') & (df_laptops['Price']>120000)
# find apple laptops that are over 120 000
df_laptops[(df_laptops['Company'] == 'Apple') & (df_laptops['Price']>120000)]
```

```python
# and -> &
# or -> |
```

```python
#------------------ Find laptops Apple or Dell laptops
#df_laptops['Company'] == 'Apple'
#df_laptops['Company'] == 'HP'
#-------------- condition
#(df_laptops['Company'] == 'Apple') | (df_laptops['Company'] == 'HP')

df_laptops[(df_laptops['Company'] == 'Apple') | (df_laptops['Company'] == 'HP')].value_counts('Company')
```

```python
# get laptops and dell laptops that cost more than 120 000
df_laptops[((df_laptops['Company'] == 'Apple') | (df_laptops['Company'] == 'HP')) & (df_laptops['Price']>120000)].value_counts('Company')
```

```python
#---------Creating a conditional column which has more than 2 choices : np.select()----------------

conditions = [
    df_laptops['Price'] > 90000,
    (df_laptops['Price'] > 50000) & (df_laptops['Price']<= 90000),
    (df_laptops['Price'] > 30000) & (df_laptops['Price'] <= 50000),
    df_laptops['Price'] <= 30000
]

values = ['Too expensive','Expensive','Affordable','Cheap']
```

```python
# set a new column
df_laptops['Price_tiers']=np.select(conditions, values,default='Unknown')
```

```python
df_laptops['Price_tiers'].value_counts()
```
