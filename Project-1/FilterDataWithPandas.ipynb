```python
#------------------------Page Two----------------------------
import pandas as pd
import numpy as np
df_laptops = pd.read_csv("../Assets/laptop_data.csv", engine='python')
```

```python
df_laptops.head(3)
```

```python
#----------------------Filter dataframe based on a condition--------------------
# find which rows have "Apple" in the Company Column

df_laptops['Company'] == "Apple"

# this return values as boolean
```

```python
# filter dataframe based on a condition
df_laptops[df_laptops['Company'] == 'Apple']
```

```python
# get a count of apple companies
df_laptops[df_laptops['Company'] == 'Apple'].value_counts('Company')
```

```python
# which rows dont have "HP"
df_laptops[df_laptops['Company'] != 'HP'].value_counts('Company')
```

```python
# find laptops over 120 000
df_laptops[df_laptops['Price']>120000].value_counts('Company')
```

```python
#------------------------Create a column based on one condition (Where)-----------------------
# create an array based on price tiers, like if condition here
np.where(df_laptops['Price']>120000,'Expensive','Cheap')
```

```python
# set a new column
df_laptops['Price_Tier'] = np.where(df_laptops['Price']>120000,'Expensive','Cheap')
```

```python
# show first five rows
df_laptops.head(5)
```

```python
# get a count in price tier column
df_laptops.value_counts('Price_Tier')
```

```python
#----------------Filter dataframe based on two conditions (And and Or Operators)------------------
# condition
(df_laptops['Company'] == 'Apple') & (df_laptops['Price']>120000)
# find apple laptops that are over 120 000
df_laptops[(df_laptops['Company'] == 'Apple') & (df_laptops['Price']>120000)]
```

```python
# and -> &
# or -> |
```

```python
#------------------ Find laptops Apple or Dell laptops
#df_laptops['Company'] == 'Apple'
#df_laptops['Company'] == 'HP'
#-------------- condition
#(df_laptops['Company'] == 'Apple') | (df_laptops['Company'] == 'HP')

df_laptops[(df_laptops['Company'] == 'Apple') | (df_laptops['Company'] == 'HP')].value_counts('Company')
```

```python
# get laptops and dell laptops that cost more than 120 000
df_laptops[((df_laptops['Company'] == 'Apple') | (df_laptops['Company'] == 'HP')) & (df_laptops['Price']>120000)].value_counts('Company')
```

```python
#---------Creating a conditional column which has more than 2 choices : np.select()----------------

conditions = [
    df_laptops['Price'] > 90000,
    (df_laptops['Price'] > 50000) & (df_laptops['Price']<= 90000),
    (df_laptops['Price'] > 30000) & (df_laptops['Price'] <= 50000),
    df_laptops['Price'] <= 30000
]

values = ['Too expensive','Expensive','Affordable','Cheap']
```

```python
# set a new column
df_laptops['Price_tiers']=np.select(conditions, values,default='Unknown')
```

```python
df_laptops['Price_tiers'].value_counts()
```

```python
#----------------isin()-Single Filtering------------------------------------------

#-select
df_laptops['Company'].isin(['Apple','HP'])
```

```python
#-filter df
df_laptops[df_laptops['Company'].isin(['Apple','HP'])
].value_counts('Company')
```

```python
#------------------isin()-Multiple Filtering----------------------
#find notebooks or ultrabooks from Apple or HP
filter1 = df_laptops['TypeName'].isin(['Notebook','Ultrabook'])
filter2 = df_laptops['Company'].isin(['Apple','HP'])

```

```python
#--filter dataframe
df_laptops[filter1 & filter2]
```

```python
#------------------Duplicate()-------------------------------
#find duplicates in one column
df_laptops.duplicated('laptop_ID')
```

```python
df_laptops[df_laptops.duplicated('laptop_ID')
]
```

```python
#find duplicates in two or more columns
# keep argument is default
df_laptops[df_laptops.duplicated(['Company','TypeName','Inches'])].sort_values(['Company','Inches'])
```

```python
# sort df ascending by company and price
df_laptops = df_laptops.sort_values(['Company','Price'])
df_laptops
```

```python
# check all the value categories in Company column
df_laptops.value_counts('Company')
```

```python
# find duplicated values in company column (default -> keep= 'first'-Keep the first duplicate value from duplicates and remove the last ones
duplicated_first = df_laptops.duplicated('Company',keep='first')
```

```python
# show df with duplicated values in company column
df_laptops[duplicated_first]

# show df with non duplicated values in Company Column (~)
df_laptops[~duplicated_first][['Company','Price']]
```

```python
df_laptops[~duplicated_first].value_counts('Company')
```

```python
# ---------------Keep =  'last' (Keep the last duplicate and remove the first ones)----------------------
duplicated_last = df_laptops.duplicated('Company',keep='last')
```

```python
df_laptops[~duplicated_last][['Company','Price']]
```

```python
#-----------------Keep = 'False' (no duplicates)-----------------------------------
duplicated_false = df_laptops.duplicated('Company',keep=False)
```

```python
df_laptops[~duplicated_false]
```

```python
df_laptops.value_counts('Company')
```
