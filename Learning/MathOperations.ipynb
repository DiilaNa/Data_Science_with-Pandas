```python
import pandas as pd
df_products = pd.read_csv("../Assets/products-1000.csv", sep=None, engine="python")

#______________________Math Operations in Columns_________________________________
```

```python
df_products
```

```python
#- select a column and calculate total
df_products['Price'].sum()
```

```python
#-calculate count 1000 - rows
df_products['Price'].count()
#-calculate average
df_products['Price'].mean()
#-calculate max number
df_products['Price'].max()
#-calculate min number
df_products['Price'].min()
#-calculate standing deviation
df_products['Price'].std()
```

```python
#- easier calculations with describe()
df_products.describe()
```

```python
#----------------------------- Math Operations in Rows ----------------------------

# - calculate sum of the rows
df_products['Price'] + df_products['Stock']

```

```python
# - calculate the average of the rows
df_products['Average'] = df_products['Price'] + df_products['Stock']/2
```

```python
df_products.round(2) #-round the decimals in to 2
```

```python
#-----------------------Value Counts------------------------
df_products['Size'].count() # returns the count of the rows
df_products['Size'].value_counts() # return the actual counts of values by category name
df_products['Size'].value_counts(normalize=True) # get as a Percentage
df_products['Size'].value_counts(normalize=True).round(2)
```

```python
#------------------Sort a Data Frame-----------------------------------
#- sort by one column - Ascending
df_products.sort_values(by=['Price'])
#- sort by one column - Descending
df_products.sort_values(by=['Price'], ascending=False)
```

```python
#- sort by multiple columns - Ascending
df_products.sort_values(['Price','Stock'])
#- sort by multiple columns - Descending
df_products.sort_values(['Price','Stock'],ascending=False)
```

```python
#- sort by multiple columns - Descending and update the original data frame
df_products.sort_values(['Price','Stock'],ascending=False,inplace=True)
```

```python
#-sort Ascending with a key function
df_products.sort_values('Name', ascending=True,key=lambda x:x.str.lower())
```
