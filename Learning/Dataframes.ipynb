```python
#----------------------Page One ---------------------------------------------------
import pandas as pd
import numpy as np
from pandas import RangeIndex

# Creating data frames - {array , dictionaries , csv files}
```

```python
# -----Option 1 -creating a dataframe using a  array-------

# creating an array
data = np.array([[1,4],[2,5],[3,6]])
# creating a data frame
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'],
                    columns=['col1', 'col2'])
# printing
df
```

```python
# ------option 2 - creating a df from a Dictionary------

state = ['California','Texas','Florida','New York']
population = [10000 , 20000,30000,40000]

dict_states = {'States':state, 'Population':population}

df_population = pd.DataFrame(dict_states)

df_population
```
