```python
import random
import pandas as pd
import numpy as np
from pandas.core.interchange.dataframe_protocol import Column

df_products = pd.read_csv("../Assets/products-1000.csv", sep=None, engine="python")
```

```python
#- creating non repetitive values for the index
new_index = np.arange(0,1000)
```

```python
#- shuffling the indexes
random.shuffle(new_index)
```

```python
#- create a new column with new index
df_products['New_Index'] = new_index
df_products
```

```python
#- set new Index
df_products.set_index('New_Index',inplace=True) # make the original one update
```

```python
#- Sort By Index
df_products.sort_index()
```

```python
#------------------Rename a column Name -------------------------------------

# - Rename a column name and overwrite the values
df_products.rename(columns={'Name':'name'})
```

```python
# - Rename multiple columns name and overwrite the values
df_products.rename(columns={'Description':'description','Brand':'brand'},inplace=True)
```
